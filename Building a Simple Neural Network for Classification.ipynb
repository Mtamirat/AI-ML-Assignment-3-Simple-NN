```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
x_train_normalized = x_train / 255.0
x_test_normalized = x_test / 255.0
```

```python
x_train_reshaped = x_train_normalized.reshape(x_train_normalized.shape[0], 28 * 28)
x_test_reshaped = x_test_normalized.reshape(x_test_normalized.shape[0], 28 * 28)
```

```python
y_train_encoded = tf.keras.utils.to_categorical(y_train, num_classes=10)
y_test_encoded = tf.keras.utils.to_categorical(y_test, num_classes=10)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation='relu', name='layer1'),
    tf.keras.layers.Dense(64, activation='relu', name='layer2'),
    tf.keras.layers.Dense(10, activation='softmax', name='layer_output')
])
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    x_train_reshaped, y_train_encoded,
    epochs=532,            
    batch_size=32,
    validation_split=0.1,
    verbose=1
)
```

```python
test_loss, test_accuracy = model.evaluate(x_test_reshaped, y_test_encoded, verbose=0)

print(f"\n✅ Test Loss: {test_loss:.4f}")
print(f"✅ Test Accuracy: {test_accuracy:.4f}")
```
